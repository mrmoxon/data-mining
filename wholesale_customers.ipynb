```python
data = pd.read_csv('./wholesale_customers.csv')
data = data.drop(['Channel', 'Region'], axis=1)
data
```

```python
describe_table = data.describe().round(0)
# print(describe_table)

statistics = {
    'min': data.min().round(0),
    'max': data.max().round(0),
    'std': data.std().round(0),
    'mean': data.mean().round(0)
}

# Create a DataFrame to display the statistics
statistics_df = pd.DataFrame(statistics)

print(statistics_df)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Standardize the dataset
scaler = StandardScaler()
data_standardized = scaler.fit_transform(data)

ks = [3, 5, 10]

best_silhouette_score = float('-inf')
best_kmeans_clusters = None
best_agglomerative_clusters = None

for k in ks:
    for i in range(10):

        # Perform k-means clustering with standardized data
        kmeans = KMeans(n_clusters=k, random_state=i)
        kmeans_clusters = kmeans.fit_predict(data_standardized)

        # Perform agglomerative hierarchical clustering with standardized data
        agglomerative = AgglomerativeClustering(n_clusters=k)
        agglomerative_clusters = agglomerative.fit_predict(data_standardized)
        
        # Compute Silhouette score for k-means clustering
        kmeans_silhouette_score = silhouette_score(data_standardized, kmeans_clusters)
        # Compute Silhouette score for agglomerative hierarchical clustering
        agglomerative_silhouette_score = silhouette_score(data_standardized, agglomerative_clusters)

        print(f"Iteration {i + 1}:")
        print("K-means Silhouette Score:", kmeans_silhouette_score)
        print("Agglomerative Hierarchical Silhouette Score:", agglomerative_silhouette_score)
        print("\n")

        # Check if the current run has the best Silhouette score for k-means clustering
        if kmeans_silhouette_score > best_silhouette_score:
            best_silhouette_score = kmeans_silhouette_score
            best_kmeans_clusters = kmeans_clusters

        # Check if the current run has the best Silhouette score for agglomerative hierarchical clustering
        if agglomerative_silhouette_score > best_silhouette_score:
            best_silhouette_score = agglomerative_silhouette_score
            best_agglomerative_clusters = agglomerative_clusters

print("Best Silhouette Score:", best_silhouette_score)
print("Best K-means Clusters:")
print(best_kmeans_clusters)
print("Best Agglomerative Hierarchical Clusters:")
print(best_agglomerative_clusters)
```
